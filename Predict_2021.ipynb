# Predict the next MVP for 2021 with random forest

You just have to follow this notebook in order to predict the ranking for 2021 data

## Import the modules needed

```python
import joblib
import numpy as np
import pandas as pd
from sklearn.ensemble import RandomForestRegressor
```

## Import and clean the data

```python
# Import
test_df = pd.read_csv('./data/test_2021.csv')
test_df.describe()

#Check for NA
print(test_df.isnull().sum().sum())
test_df.fillna(0, inplace=True)
print(test_df.isnull().sum().sum())

#Format for RF prediction
X_test = test_df.drop(columns=['player', 'year', 'team', 'pos', 'tm']) #do not account categorical var
```

## Import the model and predict

```python
rf_model = joblib.load('best_rf_model.jlib')
```

```python
test_df['pred_share'] = rf_model.predict(X_test)

test_df['rank'] = test_df['pred_share'].rank(method='min', ascending = 0)
test_df.sort_values(by=['pred_share'], ascending=False, inplace=True)
test_df
```

```python
test_df[['player', 'rank']].to_csv('Ranked_players_2021.csv', index=False)
```
